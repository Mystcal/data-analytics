Plot a side-by-side bar chart base on the data above 

The chart presents vaccination and the total of people in each country

y axis (vertical): number of people and the population

x axis (horizontal): Countries

In the graph, in green is the population, blue is the total vaccinations which had divided by 10^3 and in orange is the people fully vaccinated which also had divided by 10^4

```python
fig1,ax = plt.subplots(figsize=(8,10))

ax.set_xticks(np.arange(len(df1['country'])))
ax.set_xticklabels(df1['country'], rotation = 45)

ax.bar(df1.index+0.0, df1['total_vaccinations']/10**3,width=0.1,label='total_vaccinations(10^3)')
ax.bar(df1.index+0.1, df1['people_fully_vaccinated']/10**4,width=0.1,label='people_fully_vaccinated(10^4)')

plt.grid(True)
plt.legend()
plt.xlabel('country')
plt.ylabel('number of people')
plt.title('vaccination')

ax2 = ax.twinx()
ax2.bar(df1.index+0.2, df1['Population'],width=0.1,label='Population', color='g')
plt.legend(loc=3)
plt.ylabel('Population')
```

Filter the data needed (date, country, daily_vaccinations) from the data named (Vac) previously and only data in Australia.

```python
cy = ['Australia']
AUS = Vac[['date','country','daily_vaccinations']]
AUS = AUS[AUS.country.isin(cy)]
AUS.reset_index(inplace=True)
AUS
```

Remove the index column which it is useless and let the data frame more tidy

```python
AUS1 = AUS.drop(columns=['index'])
AUS1
```

From the date extract the month to let us easy to plot the graph

```python
AUS1['Month'] = pd.DatetimeIndex(AUS1['date']).month
AUS1
```

Then remove the date and country column which is is useless when plotting the graph

```python
AUS3 = AUS1.drop(columns=['date'])
AUS3 = AUS3.drop(columns=['country'])
AUS3
```

Sum the total vaccinations by month

```python
AUS4 = AUS3.drop_duplicates(subset='Month')
AUS4
```

remove the 'Month' column as we just need the total vaccinations to plot the graph

```python
AUS5 = AUS4.drop(columns=['Month'])
AUS5
```

Plot a line graph base on the data above 

The graph presents the total vaccinations by time

y axis (vertical): number of people

x axis (horizontal): Months

```python
AUS5.rename(columns={'daily_vaccinations':'total vaccinations'},
                    inplace = True)
LINE = AUS5.plot.line()
LINE.set_title('People get vaccinated in Austalia')
LINE.set_xlabel('Months')
LINE.set_ylabel('number of people')
LINE.set_xticklabels( [1,2,3,4,5,6,7,8] )
```
